```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8')
```

```python
df['Punto 2'] = df['n'].apply(lambda x:(((x*101**2) + (2*101) + (23))-((x*100**2) + (2*100) + (23)))/(101-100))
```

```python
def calificacion2(x):
    n=int(x[2])
    n1=int(x[7])
    if n!=n1:
        return 1
    else:
        return 5
```

```python
df["Calificacion P2"] = df.apply(calificacion2,axis=1)
```

```python
df.iloc[:, 3] = df.iloc[:, 3].astype(str).str.replace(',','.').astype(float)
```

```python
df.iloc[:, 3] = df.iloc[:, 3].astype(float)
```

```python
df['Punto 3'] = df['n'].apply(lambda x:(((x*100*2) + (2)))/(101-100))
```

```python
def calificacion3(x):
    pri = int(x[3])
    seg = int(x[9])
    if pri != seg:
        return 1
    else:
        return 5
```

```python
df['Calificacion P3'] = df.apply(calificacion3,axis = 1)
```

```python
def desviacion(nombre_archivo, nombre_columna):
    try:
        if nombre_archivo.endswith(".csv"):
            df = pd.read_csv(nombre_archivo)
        elif nombre_archivo.endswith(".xlsx") or nombre_archivo.endswith(".xls"):
            df = pd.read_excel(nombre_archivo)
        else:
            print("Formato de archivo no soportado.")
            return None
        
        if nombre_columna not in df.columns:
            print("No existe la columna")
            return None
        
        return df[nombre_columna].std()

    except FileNotFoundError:
        print("No se encontró el archivo.")
        return None
```

```python
import matplotlib.pyplot as plt

def desviacion(nombre_archivo, nombre_columna):
    try:
        if nombre_archivo.endswith(".csv"):
            df = pd.read_csv(nombre_archivo)
        elif nombre_archivo.endswith(".xlsx") or nombre_archivo.endswith(".xls"):
            df = pd.read_excel(nombre_archivo)
        else:
            print("Formato de archivo no soportado.")
            return None
        
        if nombre_columna not in df.columns:
            print("No existe la columna")
            return None

        return df[nombre_columna].std()

    except FileNotFoundError:
        print("No se encontró el archivo.")
        return None


def grafico_desviaciones(nombre_archivo):
    columnas = ["CP1", "CP2", "CP3", "NF"]
    desv = []

    # Calcular desviaciones
    for col in columnas:
        d = desviacion(nombre_archivo, col)
        desv.append(d if d is not None else 0)

    plt.figure(figsize=(8,5))
    plt.bar(columnas, desv, color=["#4C72B0", "#55A868", "#C44E52", "#8172B2"])
    plt.title("Desviaciones")
    plt.ylabel("Desviación estándar")
    plt.xlabel("Columnas")
    plt.grid(axis="y", linestyle="--", alpha=0.6)

    plt.savefig("Desviaciones.png") 
    plt.show()
```

```python
grafico_desviaciones("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```
